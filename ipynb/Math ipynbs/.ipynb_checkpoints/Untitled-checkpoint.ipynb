```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,5)
```

```python
n = [1]
dt = 1e-5
t = 0
alpha = 2
for i in range(round(1 / dt)):
    t += dt
    n.append(n[-1] * (1 + alpha * dt))
```

```python
plt.plot(n)
```
