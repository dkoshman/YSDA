```python
# -*- coding: utf-8 -*-
"""change 'poll' to 'rules' in variable names
utilize <enter> as a next entry
add relative screen entry
check if parameter rules have right syntax"""
import sys
import tkinter as tk
import numpy as np
import json
import ast

rules = {
    "poll_id" : 0,
    "screens" : []
}

list_of_screens_with_rules =[]

class MainWindow:
    """main window"""
    def __init__(self, master):
        self.master = master
        self.frame = tk.Frame(self.master)
        self.text=tk.Text(self.frame)
        self.text.insert(tk.INSERT, 'Info:')
        self.newPoll_Button = tk.Button(self.frame, text='New Rules', command = self.new_poll)
        self.poll_namelabel = tk.Label(self.frame, text='rules_name(without .json, defaulf =poll_rules)')
        self.editPoll_Button = tk.Button(self.frame, text='Edit Rules', command = self.edit_poll)
        self.poll_id_label = tk.Label(self.frame, text='poll_id:')
        self.poll_id_entry = tk.Entry(self.frame)
        self.poll_id_entry.insert(0, '0')
        self.poll_name_entry = tk.Entry(self.frame)
        self.quit_Button = tk.Button(self.frame, text='Quit', command = self.close)
        self.text.pack()
        self.poll_name_entry.pack()
        self.poll_namelabel.pack()
        self.newPoll_Button.pack()
        self.editPoll_Button.pack() 
        self.poll_id_label.pack()
        self.poll_id_entry.pack()
        self.quit_Button.pack()
        self.frame.pack()
        self.master.geometry("550x750")
        
        
    def new_poll(self):
        global poll_name, rules
        poll_name = self.poll_name_entry.get()
        if poll_name == '':
            poll_name ='poll_rules'
        if int(self.poll_id_entry.get()) !=0:
            rules['poll_id'] = int(self.poll_id_entry.get())
        self.newWindow = tk.Toplevel(self.master)
        self.app = Rules_creation_window(self.newWindow)
    
    def edit_poll(self):
        global poll_name, rules
        poll_name = self.poll_name_entry.get()
        if poll_name == '':
            poll_name ='poll_rules'
        if int(self.poll_id_entry.get()) !=0:
            rules['poll_id'] = int(self.poll_id_entry.get())
        self.newWindow = tk.Toplevel(self.master)
        self.app = Rules_edit_window(self.newWindow)
    
    def close(self):
        self.master.destroy()
        
class Rules_creation_window:
    """start creating rules"""
    def __init__(self, master):
        
        self.master = master
        self.frame = tk.Frame(self.master)
        
        
        self.new_screenLabel =tk.Label(self.frame, text='Rules will be checked on these screens(sep=\',\'):')
        self.screen_num_entry = tk.Entry(self.frame)
        #later can change to normal format, not right to left reading
        self.new_logicLabel =tk.Label(self.frame, text='Logic between rules, example: not(or(and(1,2),3))')
        self.logic_entry = tk.Entry(self.frame)
        self.logic_entry.insert(0, '1')
        self.add_screenButton=tk.Button(self.frame, text='Create rules', command=self.add_screen)
        self.quitButton = tk.Button(self.frame, text = 'Accept and save', width = 25, command = self.close_window)
        
        """packing"""
        self.new_screenLabel.pack()
        self.screen_num_entry.pack()
        self.new_logicLabel.pack()
        self.logic_entry.pack()
        self.add_screenButton.pack()
        self.quitButton.pack()
        self.frame.pack()
        self.master.geometry("500x700")
        
    def add_screen(self):
        """add window to create rules an a new screen"""
        rule = self.logic_entry.get()
        digit=0
        for i in rule:
            if i.isnumeric():
                digit+=1
        
        self.simpleCond_list=[] #list for 1,2,3 simple conditions to insert in logic entry
        
        self.newWindow = tk.Toplevel(self.master)
        self.app = Simple_condition_creation(self.newWindow, digit, self)
                
                
        
    def close_window(self):
        """check here for right encoding
        
        use json.dump instead?  """
        
        
        poll_str = str(rules)
        poll_str = poll_str.replace("True", "true" )
        poll_str = poll_str.replace("False", "false")
        poll_str = poll_str.replace("'", "\"" )
        poll_full_name = poll_name + ".json"
        open(poll_full_name, 'w').close()
        outfile = open(poll_full_name, 'r+',encoding='UTF-8')
        outfile.write(poll_str)
        outfile.close()
        
        self.master.destroy()
        
class Simple_condition_creation:
    def __init__(self, master, number, supermaster):
        self.supermaster=supermaster #master=current window? supermaster=previous window?
        self.number=number
        self.master = master
        self.toggle_check=0
        self.frame = tk.Frame(self.master)
        self.left_op = tk.IntVar()

        self.IfFirst=True
        
        self.left_opLabel= tk.Label(self.frame, text='Check for:')
        self.left_opAnswer = tk.Radiobutton(self.frame, background='orange', text = "Answer",variable = self.left_op, value = 1)
        self.left_opParameter = tk.Radiobutton(self.frame, background='orange', text = "Parameter",variable = self.left_op, value = 2)
        self.left_opAnswer.select()
        self.left_op.trace('w', self.toggle_left_op)
         

        self.left_opScreenLabel=tk.Label(self.frame, text='screen value=')
        self.left_opScreen=tk.Entry(self.frame)
        self.left_opQuesLabel=tk.Label(self.frame, text='question_id=')
        self.left_opQuesEntry=tk.Entry(self.frame)
        
        self.left_opLabel.pack()
        self.left_opAnswer.pack()
        self.left_opParameter.pack()
        
        self.simpleOpType =tk.IntVar()
        self.simpleOpEq = tk.Radiobutton(self.frame, background='pink', text = "equal",variable = self.simpleOpType, value = 1)
        self.simpleOpIn = tk.Radiobutton(self.frame, background='pink', text = "in",variable = self.simpleOpType, value = 2)
        self.simpleOpMore = tk.Radiobutton(self.frame, background='pink', text = "more",variable = self.simpleOpType, value = 3)
        self.simpleOpLess= tk.Radiobutton(self.frame, background='pink', text = "less",variable = self.simpleOpType, value = 4)
        self.simpleOpEq.select()
        
        self.onlyVar = tk.IntVar()
        
        self.onlyfalse = tk.Radiobutton(self.frame, background='yellow', text = "only=false",variable = self.onlyVar, value = 1)
        self.onlytrue = tk.Radiobutton(self.frame, background='yellow', text = "only=True",variable = self.onlyVar, value = 2)
        self.onlyfalse.select()
        
        self.constOrlistLabel =tk.Label(self.frame, text='const or list:')
        self.constOrlist = tk.Entry(self.frame)
        

        self.AcceptButton=tk.Button(self.frame, text='Accept', command=self.acceptValues)

        self.left_opScreenLabel.pack()
        self.left_opScreen.pack()
        self.left_opQuesLabel.pack()
        self.left_opQuesEntry.pack()
        self.simpleOpEq.pack()
        self.simpleOpIn.pack()
        self.simpleOpMore.pack()
        self.simpleOpLess.pack()
        self.onlyfalse.pack()
        self.onlytrue.pack()
        self.constOrlistLabel.pack()
        self.constOrlist.pack()
        

        self.AcceptButton.pack()
        
        self.frame.pack()
        self.master.geometry("500x700")
        
    def toggle_left_op(self, *args):
        
        if self.toggle_check!=2 and self.left_op.get() == 2:
            self.toggle_check=2
            self.left_opLabelParam=tk.Label(self.frame, text='parameter_id=')
            self.left_opEntryParam=tk.Entry(self.frame)
            self.left_opLabelParam.pack()
            self.left_opEntryParam.pack()

            self.simpleOpEq.pack_forget()
            self.simpleOpIn.pack_forget()
            self.simpleOpMore.pack_forget()
            self.simpleOpLess.pack_forget()
            self.onlyfalse.pack_forget()
            self.onlytrue.pack_forget()
            self.constOrlistLabel.pack_forget()
            self.constOrlist.pack_forget()
            self.AcceptButton.pack_forget()

            self.simpleOpEq.pack()
            self.simpleOpIn.pack()
            self.simpleOpMore.pack()
            self.simpleOpLess.pack()
            self.onlyfalse.pack()
            self.onlytrue.pack()
            self.constOrlistLabel.pack()
            self.constOrlist.pack()
            self.AcceptButton.pack()
            
            try:
                self.left_opScreenLabel.pack_forget()
                self.left_opScreen.pack_forget()
                self.left_opQuesLabel.pack_forget()
                self.left_opQuesEntry.pack_forget()
            except:
                pass
     
            self.frame.pack()

            
        if self.toggle_check!=1 and self.left_op.get() == 1:
            self.toggle_check=1
            self.left_opScreenLabel=tk.Label(self.frame, text='screen_id=')
            self.left_opScreen=tk.Entry(self.frame)
            self.left_opQuesLabel=tk.Label(self.frame, text='question_id=')
            self.left_opQuesEntry=tk.Entry(self.frame)
            self.left_opScreenLabel.pack()
            self.left_opScreen.pack()
            self.left_opQuesLabel.pack()
            self.left_opQuesEntry.pack()
            

            self.simpleOpEq.pack_forget()
            self.simpleOpIn.pack_forget()
            self.simpleOpMore.pack_forget()
            self.simpleOpLess.pack_forget()
            self.onlyfalse.pack_forget()
            self.onlytrue.pack_forget()
            self.constOrlistLabel.pack_forget()
            self.constOrlist.pack_forget()
            self.AcceptButton.pack_forget()
            
            self.simpleOpEq.pack()
            self.simpleOpIn.pack()
            self.simpleOpMore.pack()
            self.simpleOpLess.pack()
            self.onlyfalse.pack()
            self.onlytrue.pack()
            self.constOrlistLabel.pack()
            self.constOrlist.pack()
            self.AcceptButton.pack()
            
            self.left_opEntryParam.pack_forget()
            self.left_opLabelParam.pack_forget()
   
            self.frame.pack()
        
    def acceptValues(self):
        """clear window, save values, check if over ( digits value=windows created )"""
        global rules
        
        condition = {
            "operator": "equal",
            "only": False,
            "left_operand": {
              "type": "question",
              "property": "answer_id"
            },
            "right_operand": {
              "type": "constant",
              "property": 1
            }
          }
        #screens_tolook= [int(i) for i in self.left_opScreen.get().split(',')]
        if self.left_op.get() == 1: #type=answer
            condition['left_operand']['screen_id']=int(self.left_opScreen.get())
            condition['left_operand']['question_id'] =int(self.left_opQuesEntry.get())
        else:
            condition['left_operand']['type'] ='parameter'
            condition['left_operand']['property'] = int(self.left_opEntryParam.get())  #set parameter id
        
        if self.onlyVar.get()==2:
            condition['only']=True
            
        if self.simpleOpType.get() ==2:
            condition['operator'] ='in'
            condition['right_operand']['type'] = 'list'
        elif self.simpleOpType.get() ==3:
            condition['operator'] = 'more'
        elif self.simpleOpType.get() ==4:
            condition['operator'] ='less'
        
        right_op_prop =[int(i) for i in self.constOrlist.get().split(',')]
        if condition['operator']=='in':
            condition['right_operand']['property']= right_op_prop
        else:
            condition['right_operand']['property']=right_op_prop[0]
        
        print(condition)
        self.supermaster.simpleCond_list.append(condition)
            
        
        if self.number>1:
            self.newWindow = tk.Toplevel(self.master.master)
            self.app = Simple_condition_creation(self.newWindow, self.number-1, self.supermaster)
            
        else: 
            self.newWindow = tk.Toplevel(self.master.master)
            self.app = Action_creation_window(self.newWindow, self.supermaster)

            
        self.master.destroy()
        

class Action_creation_window:
    """create action after accepting all simple conditions"""
    def __init__(self, master, supermaster):
        self.master = master
        self.supermaster=supermaster
        self.frame = tk.Frame(self.master)
        self.screen_list = [int(i) for i in str(self.supermaster.screen_num_entry.get()).split(',')]
        
        self.actionVar=tk.IntVar()
        self.actionLabel=tk.Label(self.frame, text='action type')
        self.actionScreen=tk.Radiobutton(self.frame, background='orange', text = "add_screen",variable = self.actionVar, value = 1)
        self.actionScreenList=tk.Radiobutton(self.frame, background='orange', text = "add_screen_list",variable = self.actionVar, value = 2)
        self.actionQuestion=tk.Radiobutton(self.frame, background='orange', text = "add_question",variable = self.actionVar, value = 3)
        self.actionParameter=tk.Radiobutton(self.frame, background='orange', text = "increase parameter",variable = self.actionVar, value = 4)
        self.actionVar.trace('w', self.add_ques_default)
        
        self.actionScreen.select()
        self.actionEntryLabel1=tk.Label(self.frame, text='screen/parameter')
        self.actionEntry1 =tk.Entry(self.frame)
        self.actionEntryLabel2=tk.Label(self.frame, text='question/parameter increase')
        self.actionEntry2 =tk.Entry(self.frame)
        
        self.acceptButton =tk.Button(self.frame, text='Accept', command=self.accept_values)
        
        self.actionLabel.pack()
        self.actionScreen.pack()
        self.actionScreenList.pack()
        self.actionQuestion.pack()
        self.actionParameter.pack()
        self.actionEntryLabel1.pack()
        self.actionEntry1.pack()
        self.actionEntryLabel2.pack()
        self.actionEntry2.pack()
        self.acceptButton.pack()
        
        self.frame.pack()
        self.master.geometry('500x700')
        
    def add_ques_default(self, *args):
        if self.actionVar.get()==3:
            self.actionEntry1.insert(0,self.screen_list[0])
            self.actionEntry2.insert(0, 2)
    
    def accept_values(self):
        '''main processing function'''
        global rules
        
        screen_rule={}
        
        rule=str(self.supermaster.logic_entry.get())
        rule=rule.replace('(', '')
        rule=rule.replace('or',    str("{\"operator\": \"or \",\"conditions\":["))
        rule =rule.replace('not',  str("{\"operator\": \"not\",\"conditions\":["))
        rule =rule.replace('and',  str("{\"operator\": \"and\",\"conditions\":["))
        rule=rule.replace(')', ']}')
        rule ='{ \"condition\":' + rule+'}'
        
        for i in range(len(self.supermaster.simpleCond_list)):
            rule=rule.replace(str(1+i), str(self.supermaster.simpleCond_list[i]))
        rule_json = ast.literal_eval(rule)
        
        if self.actionVar.get()==1:
            rule_json['action_type']='add_screen'
            rule_json['object'] ={ 'type':'screen', 'screen_id' : int(self.actionEntry1.get()) } 
        if self.actionVar.get()==2:
            rule_json['action_type']='add_screen_list'
            rule_json['object'] ={ 'type':'screen_list', 'property' :  [int(i) for i in self.actionEntry1.get().split(',')] }
        if self.actionVar.get()==3:
            rule_json['action_type']='add_question'
            rule_json['object'] ={ 'type':'question', 'screen_id' : int(self.actionEntry1.get()), 'question_id': int(self.actionEntry2.get()) } 
        if self.actionVar.get()==4:
            rule_json['action_type']='increase'
            rule_json['object'] ={ 'type':'parameter', 'parameter_id' : int(self.actionEntry1.get()) } 
            rule_json['additional_operand'] ={ 'type':'constant', 'property' : int(self.actionEntry2.get()) } 
        
        
        for i in self.screen_list:
                   
            new = True
            for j in rules['screens']:
                new = new and not j['number']==i
                
            if new:
                rules['screens'].append({'number':i, 'rules':[rule_json]})
            else:
                rules['screens']['number'==i].append(rule_json)
                
        #comment if dont to clear values on 2nd window
        self.supermaster.logic_entry.delete(0, 'end') 
        self.supermaster.logic_entry.insert(0, '1')
        #self.supermaster.screen_num_entry.delete(0, 'end')
        self.master.destroy()
        
    
class Rules_edit_window:
    """start editing rules"""

def main(): 
    root = tk.Tk()
    root.tk.call('encoding', 'system', 'utf-8')
    app = MainWindow(root)
    root.mainloop()
    

if __name__ == '__main__':
    main()
```
