```python
import pybrain
import pandas as pd
```

```python
df=pd.read_csv("http://math-info.hse.ru/f/2016-17/nes-ml/neuro_class.csv")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df["x0"], df['x1'], 20, df['class'])
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df[ ['x0', 'x1'] ].values, df['class'].values)
```

```python
from pybrain.datasets import ClassificationDataSet
```

```python
def build_data_set(X, y):
    data = ClassificationDataSet(2, 1)
    for x, y0 in zip(X, y):
        data.addSample(x, y0)
    return data
```

```python
train = build_data_set(X_train, y_train)
```

```python
test = build_data_set(X_test, y_test)
```

```python
train.getFieldNames()
```

```python
train._convertToOneOfMany()
```

```python
train['target']
```

```python
from pybrain.tools.shortcuts import buildNetwork
```

```python
from pybrain.structure import TanhLayer, SoftmaxLayer
```

```python
network = buildNetwork(2, 10, 2, hiddenclass = TanhLayer, 
                      outclass=SoftmaxLayer)
# 2 10 2: 10 -- number of neurons
```

```python
from pybrain.supervised.trainers import BackpropTrainer
```

```python
trainer = BackpropTrainer(network, train, weightdecay=1e-5)
```

```python
trainer.trainEpochs(10)
```

```python
trainer.module.activate([1, 1])
```

```python
test._convertToOneOfMany()
```

```python
import numpy as np
test_predicted = np.argmax(trainer.module.activateOnDataset(test), axis=1)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], 20, test_predicted)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], (test_predicted + 0.4)*10)
```

```python
new_test = np.random.uniform(-10, 10, size =[1000, 2])
```

```python
new_test_data = build_data_set(new_test, np.zeros([1000, 1]))
```

```python
test_predicted = np.argmax(trainer.module.activateOnDataset(new_test_data), axis=1)
```

```python
plt.scatter(new_test[:, 0], new_test[:, 1],  (test_predicted + 0.4)*10)
```
