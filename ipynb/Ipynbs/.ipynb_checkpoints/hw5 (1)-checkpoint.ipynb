Это ДЗ посвящено решающим деревьям (decision trees) и разложению ошибок на смещение (bias) и разброс (variance). Оно частично основано на [этой практической работе](https://github.com/esokolov/ml-course-hse/blob/master/2016-fall/homeworks-practice/homework-practice-03-ensembles.ipynb). Пусть имеется функция $y=f(x)$, задающая истинную связь между $x\in \mathbb X$ и $y\in \mathbb Y$. Очередной элемент данных $(x_i, y_i)$ генерируется следующим образом: сначала выбирается случайный элемент $x_i$ (в соответствии с некоторым распределением, заданным на $\mathbb X$). Затем вычисляется $y_i = f(x_i)+\varepsilon_i$, где $\varepsilon_i$ — случайный шум, выбирающийся из ещё какого-то распределения. Повторяя эту операцию много раз, мы можем получить выборку любого заданного размера. Заметим, что свойства получающихся выборок зависят не только от функции $f$, но и от выбранных распределений.

### Задача 1
Пусть $\mathbb X=\mathbb R_+$, $f(x)=\sin(x)$, распределение на $\mathbb X$ — экспоненциальное с $\lambda=1$ (`np.random.exponential`), случайный шум выбирается из равномерного распределения на отрезке $[-1, 1]$ (`np.random.uniform` — нужно указать правильно параметры).

Написать функцию `make_sample(size)`, генерирующую выборку длины `size`. Она должна возвращать пару `(X, y)`, где `X` и `y` можно было бы использовать для обучения регрессоров `sklearn` (то есть `X` и `y` должны быть вектор-столбцами — можно использовать метод `.reshape(-1, 1)`).

```python
import numpy as np
def make_sample(size=1):
    X = np.random.exponential(1, size)
    X = X.reshape(-1, 1)
    noise = np.random.uniform(-1, 1, size)
    noise = noise.reshape(-1, 1)
    y = np.sin(X) + noise
    return(X, y)
```

### Задача 5
Написать функцию `estimate_tree_bias(max_depth)`, оценивающую смещение решающего дерева глубины `max_depth`. Найти смещения для деревьев из задач 3 и 4. 

Напомним, что смещение задаётся формулой

$$\mathbb E_{x}\left[
(\mathbb E_{\mathcal X}[\mu(\mathcal X)(x)]-\mathbb E[y|x])^2\right].
$$

Здесь $\mathcal X\in (\mathbb X\times \mathbb Y)^n$ — выборка размером $n$, $\mu(\mathcal X)$ — способ обучения алгоритма по выборке, $\mu(\mathcal X)(x)$ — предсказанное алгоритмом значение на объекте $x\in \mathbb X$, $\mathbb E[y|x]$ — условное матожидание $y$ при условии $x$ (для нашего алгоритма генерирования данных), $\mathbb E_{x}$ берётся в соответствии с вероятностным распределением, заданным на $\mathbb X$, $\mathbb E_{\mathcal X}$ берётся в соответствии с вероятностным распределением, заданным на выборках (опять же, в соответствии с нашим способом генерировать данные).

Таким образом, для оценки смещения вам необходимо:

1. Сгенерировать случайный $x$ в соответствии с выбранным распределением.
2. Сгенерировать много (например, 100) выборок $\mathcal X$ одинаковой длины (например, 300). Обучить по каждой из них алгоритм. Найти среднее предсказанное значение в точке $x$. Найти разность между средним предсказанным значением и $\mathbb E[y|x]$ (подсказка: в нашем случае это просто $f(x)$, т.к. матожидание шума равно нулю). Возвести в квадрат. Записать.
3. Повторить шаги 1-2 много (например, 300) раз. Усреднить.

Варить до готовности, посолить по вкусу.

Смещение готово!

(Хозяйке на заметку: чтобы программа работала быстрее, можно сразу сгенерировать много `x`'ов, а потом для каждой выборки генерировать сразу много предсказаний на всех `x`'ах сразу. Но можно обойтись и без этого.)

```python
def estimate_tree_bias(max_tree_depth):
    sample_size = 300
    num_regressions = 5000
    X, y = make_sample(sample_size)
    average_prediction = np.array([0]*sample_size)
    average_prediction = average_prediction.reshape(-1, 1)
    for i in range(num_regressions):
        Xi, yi = make_sample(sample_size)
        regressor = DecisionTreeRegressor(max_depth = max_tree_depth)
        regressor.fit(Xi, yi)
        predictions = regressor.predict(X)
        predictions = predictions.reshape(-1, 1)
        average_prediction = average_prediction + predictions 
    
    average_prediction = average_prediction/num_regressions
    bias_matrix = (average_prediction - np.sin(X))**2
    bias=sum(bias_matrix)/sample_size
    return (bias)
```

```python
tree_2 = estimate_tree_bias(2)
tree_4 = estimate_tree_bias(4)
tree = estimate_tree_bias(None)
print( "bias with tree depth 2: ", tree_2, "\n", "bias with tree depth 4: ", tree_4, "\n", "bias with default tree depth:", tree)
```

### Задача 6

Написать функцию `estimate_tree_variance(max_depth)`, оценивающую разброс решающего дерева глубины `max_depth`. Найти разброс для деревьев из задач 3 и 4.

Напомним, что разброс задаётся формулой

$$\mathbb E_{x}\left[\mathbb E_{\mathcal X}
\left[(\mu(\mathcal X)(x)-\mathbb E_{\mathcal X}[\mu(\mathcal X)(x)])^2\right]
\right].$$

Таким образом, для оценки разброса вам потребуется следующее:

1. Сгенерировать случайный $x$ в соответствии с выбранным распределением.
2. Сгенерировать много выборок $\mathcal X$. Для каждой обучить алгоритм и найти его предсказание на $x$. Найти выборочную дисперсию полученных предсказаний. Записать её.
3. Повторить шаги 1 и 2 много раз. Усреднить.

Разброс готов!

```python
def estimate_tree_variance(max_tree_depth):
    sample_size = 300
    num_regressions = 5000
    
    X, y = make_sample(sample_size)
    sample_sum = np.array([0.]*sample_size)
    average_prediction = np.array([0.]*sample_size)
    quadratic = np.array([0.]*num_regressions)
    X_predictions = np.array([ [0.]*sample_size ]* num_regressions)
    
    for i in range(num_regressions):
        Xi, yi = make_sample(sample_size)
        regressor = DecisionTreeRegressor(max_depth = max_tree_depth)
        regressor.fit(Xi, yi)
        predictions = regressor.predict(X)
        X_predictions[i] = predictions
    for i in range(num_regressions):
        sample_sum += X_predictions[i]
    sample_sum = sample_sum/num_regressions
    for i in range(num_regressions):
        tmp =  (sample_sum - X_predictions[i]) ** 2
        quadratic[i] = sum(tmp)/sample_size
    variance = sum(quadratic)/num_regressions
    return (variance)   
```

```python
x2 = estimate_tree_variance(2)
x4 = estimate_tree_variance(4)
x = estimate_tree_variance(None)
print( "variance with tree depth 2: ", x2, "\n", "variance with tree depth 4: ", x4, "\n", "variance with default tree depth:", x)
```

Видно, что с увеличением глубины дерева смещение уменьшается, а дисперсия увеличивается

### Задача 7 (необязательная)
Исследовать, как изменятся смещение и разброс при изменении распределения, заданного на $\mathbb X$. Что изменится, если вместо экспоненциального распределения использовать равномерное на отрезке от $0$ до $7$? 

```python
def make_uniform_sample(size=1):
    X = np.random.uniform(0, 7, size)
    X = X.reshape(-1, 1)
    noise = np.random.uniform(-1, 1, size)
    noise = noise.reshape(-1, 1)
    y = np.sin(X) + noise
    return(X, y)

def estimate_tree_bias_uniform(max_tree_depth):
    sample_size = 300
    num_regressions = 5000
    X, y = make_uniform_sample(sample_size)
    average_prediction = np.array([0]*sample_size)
    average_prediction = average_prediction.reshape(-1, 1)
    for i in range(num_regressions):
        Xi, yi = make_uniform_sample(sample_size)
        regressor = DecisionTreeRegressor(max_depth = max_tree_depth)
        regressor.fit(Xi, yi)
        predictions = regressor.predict(X)
        predictions = predictions.reshape(-1, 1)
        average_prediction = average_prediction + predictions 
    
    average_prediction = average_prediction/num_regressions
    bias_matrix = (average_prediction - np.sin(X))**2
    bias=sum(bias_matrix)/sample_size
    return (bias)

tree_2 = estimate_tree_bias_uniform(2)
tree_4 = estimate_tree_bias_uniform(4)
tree = estimate_tree_bias_uniform(None)
print( "bias with tree depth 2: ", tree_2, "\n", "bias with tree depth 4: ", tree_4, "\n", "bias with default tree depth:", tree)

def estimate_tree_variance_uniform(max_tree_depth):
    sample_size = 300
    num_regressions = 5000
    
    X, y = make_uniform_sample(sample_size)
    sample_sum = np.array([0.]*sample_size)
    average_prediction = np.array([0.]*sample_size)
    quadratic = np.array([0.]*num_regressions)
    X_predictions = np.array([ [0.]*sample_size ]* num_regressions)
    
    for i in range(num_regressions):
        Xi, yi = make_uniform_sample(sample_size)
        regressor = DecisionTreeRegressor(max_depth = max_tree_depth)
        regressor.fit(Xi, yi)
        predictions = regressor.predict(X)
        X_predictions[i] = predictions
    for i in range(num_regressions):
        sample_sum += X_predictions[i]
    sample_sum = sample_sum/num_regressions
    for i in range(num_regressions):
        tmp =  (sample_sum - X_predictions[i]) ** 2
        quadratic[i] = sum(tmp)/sample_size
    variance = sum(quadratic)/num_regressions
    return (variance)   

x2 = estimate_tree_variance_uniform(2)
x4 = estimate_tree_variance_uniform(4)
x = estimate_tree_variance_uniform(None)
print( "variance with tree depth 2: ", x2, "\n", "variance with tree depth 4: ", x4, "\n", "variance with default tree depth:", x)
```

   При экспоненциальном распределении смещение и разброс предсказаний дерева меньше, чем при равномерном при 
одинаковой глубине дерева. Интуитивно это можно объяснить тем, что при экспоненциальном рапределении большинство
выборок находятся ближе друг к дргугу, чем при равномерном, а поскольку при большей плотности выборки строить 
приближение легче, то и смещение и разброс тоже меньше. Из-за экспоненциального распределения также есть и выбросы,
которые трудно предсказывать, но поскольку ошибка на выбросах квадратичная и растет медленней экспоненциальной, то и
в среднем эти выбросы не перевешивают выгоды экспоненциального распределения. То есть на экспоненциальной выборке
быстро учиться, но при неограниченной глубине дерева выбросы портят смещение по сравнению с равномерным 
распределением примерно на $10\%$ с одинаковой дисперсией.
