Duncan index

$$ID = \frac{1}{2} \sum |\frac{F_i}{F} - \frac{M_i}{M}|$$

```python
total_table = df[(df['indicator.label'] == 'Employment by sex and economic activity (thousands)') &
(df['classif1.label'] == 'ISIC-Rev.4: TOTAL. Total')].pivot(index='sex.label', columns='time', values='obs_value')
```

```python
IDs = []
group = 'ISIC-Rev.2'
indicator = 'Employment by sex and economic activity (thousands)'
total_table = df[(df['indicator.label'] == 'Employment by sex and economic activity (thousands)') &
(df['classif1.label'] == group + ': TOTAL. Total')].pivot(index='sex.label', columns='time', values='obs_value')

for year in total_table.columns:
    ID = 0
    F = total_table[year].loc['Sex: Female']
    M = total_table[year].loc['Sex: Male']

    for i in df[names][(df['indicator.label'] == indicator) &
    df['classif1.label'].isin(isic_dict[group])]\
    [['time', 'obs_value', 'sex.label', 'classif1.label']]\
    .groupby('classif1.label'):
        table = i[1][['time', 'obs_value', 'sex.label']].pivot(index='time', columns='sex.label')
        table.columns = [j[1] for j in table.columns]
        ID += abs(table['Sex: Female'].loc[year] / F - table['Sex: Male'].loc[year] / M)
    IDs.append(ID / 2)
    
fig, axis = plt.subplots(figsize=(10, 5))
plt.plot(total_table.columns, IDs)
_ = axis.set_title('Duncan Index by economic activity, 1980-1994')
```

Sex relation index

$$SR = \frac{N}{F} \cdot (\frac{F_f}{N_f} - \frac{F_m}{N_m})$$

```python
SRs = []
group = 'ISIC-Rev.2'
indicator = 'Employment by sex and economic activity (thousands)'
total_table = df[(df['indicator.label'] == 'Employment by sex and economic activity (thousands)') &
(df['classif1.label'] == group + ': TOTAL. Total')].pivot(index='sex.label', columns='time', values='obs_value')

for year in total_table.columns:
    
    SR = 0
    F = total_table[year].loc['Sex: Female']
    M = total_table[year].loc['Sex: Male']
    F_f = 0
    F_m = 0
    M_f = 0
    M_m = 0
    
    for i in df[names][(df['indicator.label'] == indicator) &
    df['classif1.label'].isin(isic_dict[group])]\
    [['time', 'obs_value', 'sex.label', 'classif1.label']]\
    .groupby('classif1.label'):
        table = i[1][['time', 'obs_value', 'sex.label']].pivot(index='time', columns='sex.label')
        table.columns = [j[1] for j in table.columns]
        if table['Sex: Female'].loc[year] > table['Sex: Male'].loc[year]:
            F_f += table['Sex: Female'].loc[year]
            M_f += table['Sex: Male'].loc[year]
        else:
            F_m += table['Sex: Female'].loc[year]
            M_m += table['Sex: Male'].loc[year]
    
    N_f = F_f + M_f
    N_m = F_m + M_m
    N = F + M
    SRs.append((N / F) * (F_f / N_f - F_m / N_m))
    
fig, axis = plt.subplots(figsize=(10, 5))
plt.plot(total_table.columns, SRs)
_ = axis.set_title('Sex Relation Index by economic activity, 1980-1994')
```

Women employment index

$$WE = \sum |\frac{F_i}{F} - \frac{N_i}{N}|$$

```python
WEs = []
group = 'ISIC-Rev.2'
indicator = 'Employment by sex and economic activity (thousands)'
total_table = df[(df['indicator.label'] == 'Employment by sex and economic activity (thousands)') &
(df['classif1.label'] == group + ': TOTAL. Total')].pivot(index='sex.label', columns='time', values='obs_value')

for year in total_table.columns:
    
    WE = 0
    F = total_table[year].loc['Sex: Female']
    M = total_table[year].loc['Sex: Male']
    
    for i in df[names][(df['indicator.label'] == indicator) &
    df['classif1.label'].isin(isic_dict[group])]\
    [['time', 'obs_value', 'sex.label', 'classif1.label']]\
    .groupby('classif1.label'):
        table = i[1][['time', 'obs_value', 'sex.label']].pivot(index='time', columns='sex.label')
        table.columns = [j[1] for j in table.columns]
        
        WE += abs(table['Sex: Female'].loc[year] / F - (table['Sex: Female'].loc[year] + \
                                                        table['Sex: Male'].loc[year]) / (M + F))
    
    WEs.append(WE)
    
fig, axis = plt.subplots(figsize=(10, 5))
plt.plot(total_table.columns, WEs)
_ = axis.set_title('Women Employment Index by economic activity, 1980-1994')
```

Marginal matching index

$$MM = \frac{F_f M_m - F_m M_f}{F M}$$

```python
MMs = []
group = 'ISIC-Rev.2'
indicator = 'Employment by sex and economic activity (thousands)'
total_table = df[(df['indicator.label'] == 'Employment by sex and economic activity (thousands)') &
(df['classif1.label'] == group + ': TOTAL. Total')].pivot(index='sex.label', columns='time', values='obs_value')

for year in total_table.columns:
    
    F = total_table[year].loc['Sex: Female']
    M = total_table[year].loc['Sex: Male']
    F_f = 0
    F_m = 0
    M_f = 0
    M_m = 0
    
    for i in df[names][(df['indicator.label'] == indicator) &
    df['classif1.label'].isin(isic_dict[group])]\
    [['time', 'obs_value', 'sex.label', 'classif1.label']]\
    .groupby('classif1.label'):
        table = i[1][['time', 'obs_value', 'sex.label']].pivot(index='time', columns='sex.label')
        table.columns = [j[1] for j in table.columns]
        if table['Sex: Female'].loc[year] > table['Sex: Male'].loc[year]:
            F_f += table['Sex: Female'].loc[year]
            M_f += table['Sex: Male'].loc[year]
        else:
            F_m += table['Sex: Female'].loc[year]
            M_m += table['Sex: Male'].loc[year]
    
    MMs.append((F_f * M_m - F_m * M_f) / (F * M))
    
fig, axis = plt.subplots(figsize=(10, 5))
plt.plot(total_table.columns, MMs)
_ = axis.set_title('Marginal Matching Index by economic activity, 1980-1994')
```
