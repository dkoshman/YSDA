### Markdown formatting

```python
import re


def create_link(description, section_name):
    return f"[{description}](#{section_name.replace(' ', '-')})"


def add_link(task: str):
    return re.sub(
        r"(Задача \d+) \[.*?\]",
        lambda match: "### " + create_link(match.group(0), match.group(1)),
        task,
    )


def create_solution_template(task: str) -> str:
    task = re.sub(r"(Задача \d+) \[.*?\]", r"### \1", task)
    header, *subtasks = re.split(r"(?=1\. )", task)
    subtasks = [
        subtask.replace("1. ", f"{i}. ") for i, subtask in enumerate(subtasks, 1)
    ]
        
    tex_template = r"""
$\displaystyle{
}$

$\displaystyle{
\begin{align*}
 =
&  =\\=
\end{align*}
}$
    """
    
    subtasks = ["---\n" + s + tex_template for s in subtasks]
    template = header + "".join(subtasks)
    return template


def format_tasks(text):
    tasks = re.split(r"(?=Задача \d+)", text)[1:]
    tasks_with_links = "".join(add_link(task) for task in tasks)
    solutions = "".join(create_solution_template(task) for task in tasks)

    print("\n\nFormatted output:\n")
    print(tasks_with_links)
    print("\n\n")
    print(solutions)
```

```python
format_tasks(input())
```

### Задача 1
Пусть $\boldX = \{X_1,\dots,X_n\} \sim \Normal(\theta,1)$. Пусть $\esttheta_n = a\mean{\boldX}$, где $\mean{\boldX} = n^{-1}\sum_{i=1}^n X_i$, и $0\leq a\leq 1$ –– некоторая константа. Найти $\MSE$ для оценки $\esttheta_n$. Какое значение $a$ позволяет достигнуть наименьшего значения $\MSE$? Чему равно наименьшее значение $\MSE$?

---

$\displaystyle{
\begin{align*}
\text{MSE}(\hat\theta_n) = 
& \Exp((\hat\theta_n - \theta)^2) =\\=
& \Exp((a\mean\boldX - \theta)^2) =\\=
& a^2\Exp((\mean\boldX)^2) -2a\theta\Exp(\mean\boldX) + \theta^2 =\\=
& a^2\Exp((\mean\boldX)^2) -2a\theta^2 + \theta^2 
\end{align*}
}$

$\displaystyle{
\begin{align*}
\Exp((\mean\boldX)^2) = 
& \frac{1}{n^2}\Exp\left(\sum_i X_i^2 + \sum_i\sum_{i\neq j} X_i X_j\right) =\\=
& \frac{1}{n^2}\left(n (1 + \theta^2) + n(n-1)\theta^2\right) =\\=
& \frac{1}{n}\left(1 + n\theta^2\right)
\end{align*}
}$

$\displaystyle{
\text{MSE}(\hat\theta_n) = 
a^2\frac{1}{n}\left(1 + n\theta^2\right) -2a\theta^2 + \theta^2 =
\frac{a^2}{n} + \theta^2 \left(a^2 - 2a + 1 \right) =
\frac{a^2}{n} + \theta^2 \left(a - 1\right)^2
}$

$\displaystyle{
\begin{align*}
\frac{d}{da} =
\frac{2a}{n} + 2\theta^2 \left(a - 1\right) =
2\left(a(\frac{1}{n} + \theta^2) - \theta^2\right)
\end{align*}
}$

$\displaystyle{
\begin{align*}
a' = 
\arg\min_{0\leq a\leq1}\text{MSE}(\hat\theta_n) =
\frac{n\theta^2}{n\theta^2 + 1} \in [0, 1]
\end{align*}
}$

$\displaystyle{
\begin{align*}
\text{MSE}(\hat\theta_n') =
& \frac{a'^2}{n} + \theta^2 \left(a' - 1\right)^2 =\\=
& \frac{1}{n}\left(\frac{n\theta^2}{n\theta^2 + 1}\right)^2 + \theta^2 \left(\frac{n\theta^2}{n\theta^2 + 1} - 1\right)^2 =\\=
& \frac{n\theta^4}{(n\theta^2 + 1)^2} + \frac{\theta^2}{(n\theta^2 + 1)^2} =\\=
& \frac{\theta^2}{n\theta^2 + 1}
\end{align*}
}$

### Задача 2
Пусть $\boldX =\{X_1,\ldots,X_n\} \sim \Gamma(\alpha,\beta)$ (гамма-распределение). Найдите с помощью метода моментов оценки для $\alpha$ и $\beta$. 

_Подсказка: считать известным, что $\Exp[X] = \alpha \beta$, $\Var[X] = \alpha \beta^2$._

---

Метод моментов состоит в том, чтобы взять набор функций $g_i$ от случайной величины $X$ таких, что их мат ожидание было равно нулю, и заменить мат ожидание на выборочное среднее:

$\displaystyle{
\begin{cases}
    \Exp[X] - \alpha\beta = 0 \\
    \Var[X] - \alpha \beta^2 = 0\\
\end{cases} \Leftrightarrow
\begin{cases}
    \Exp[X] - \alpha\beta = 0 \\
    \Exp[X^2] - \Exp[X]^2 - \alpha \beta^2 = 0\\
\end{cases}
\Leftrightarrow
\begin{cases}
    \Exp[X] - \alpha\beta = 0 \\
    \Exp[X^2] - \alpha^2 \beta^2 - \alpha \beta^2 = 0\\
\end{cases}
\Leftrightarrow
\begin{cases}
    \Exp[g_1(X)] = 0 \\
    \Exp[g_2(X)] = 0\\
\end{cases}
}$

Где $g_1(x) = x - \alpha\beta, g_2(x) = x^2 - \alpha^2 \beta^2 - \alpha \beta^2$. Воспользуемся методом моментов:

$\displaystyle{
\begin{cases}
    \mean{g_1(X)} = 0 \\
    \mean{g_2(X)} = 0\\
\end{cases} \Leftrightarrow
\begin{cases}
    \hat\beta = \cfrac{\mean{X}}{\hat\alpha} \\
    \hat\beta^2 = \cfrac{\mean{X^2}}{\hat\alpha^2  + \hat\alpha} \\
\end{cases} \Leftrightarrow
\begin{cases}
    \hat\beta = \cfrac{\mean{X}}{\hat\alpha} \\
    \cfrac{\mean{X}^2}{\hat\alpha^2} = \cfrac{\mean{X^2}}{\hat\alpha^2  + \hat\alpha} \\
\end{cases} \Leftrightarrow
\begin{cases}
    \hat\beta = \cfrac{\mean{X}}{\hat\alpha} \\
    \hat\alpha = \cfrac{\mean{X}^2}{\mean{X^2} - \mean{X}^2} \\
\end{cases} \Leftrightarrow
\begin{cases}
    \hat\beta = \cfrac{\mean{X^2} - \mean{X}^2}{\mean{X}} \\
    \hat\alpha = \cfrac{\mean{X}^2}{\mean{X^2} - \mean{X}^2} \\
\end{cases} 
}$

$$
p(x, \theta) = \frac{1}{2\sigma}e^{-\frac{|x - \mu|}{\sigma}}, \quad x \in \RR.
$$
Найдите MLE-оценку для $\theta$.

---

$\displaystyle{
\begin{align*}
L(\theta) =
& \prod_i p(X_i| \theta) =\\=

& \exp\left\{\sum_i \ln \frac{1}{2\sigma}\exp\left\{-\frac{|X_i - \mu|}{\sigma}\right\} \right\} =\\=
& \exp\left\{
- n\ln 2 - n \ln \sigma - 
\frac{1}{\sigma}\sum_i |X_i - \mu|
\right\}
\end{align*}
}$

$\displaystyle{
(\hat\mu, \hat\sigma) =
\arg\max_{\mu, \sigma} L(\theta)=
\arg\min_{\mu, \sigma} \left(n\ln \sigma + \frac{1}{\sigma}\sum_i |X_i - \mu|\right) =
\arg\min_{\mu, \sigma} \left(n\ln \sigma + \frac{1}{\sigma}\sum_i (X_i - \mu)(1 - 2\cdot\mathbb{1}[X_i < \mu])\right)
}$

$\displaystyle{
\frac{d}{d\sigma} = \frac{n}{\sigma} - \frac{1}{\sigma^2}\sum_i |X_i - \mu| \\
\frac{d}{d\mu} = \frac{1}{\sigma}\sum_i(2\cdot\mathbb{1}[X_i < \mu] - 1) =
\frac{2}{\sigma}\left(\sum_i\mathbb{1}[X_i < \mu] - \frac{n}{2}\right)\\
}$

$\displaystyle{
\hat\sigma = \frac{1}{n}\sum_i |X_i - \hat\mu| \\
 \\
\begin{cases}
    \sum_i\mathbb{1}[X_i < \hat\mu] = \frac{n}{2}, &n - \text{четное}\\
    \hat\mu = \boldX_{(n + 1)/2}, &n - \text{нечетное} \\
\end{cases} \Rightarrow
\begin{cases}
    \hat\mu \in (\boldX_{n/2}, \boldX_{n/2 + 1}], &n - \text{четное} \\
    \hat\mu = \boldX_{(n + 1)/2}, &n - \text{нечетное} \\
\end{cases}
}$

### Задача 4
Пусть $\boldX = \{X_1,\dots,X_n\} \sim \Uniform (\theta,2\theta)$, $\theta > 0$.

1. Найдите MLE-оценку $\mletheta$ для параметра $\theta$. Найдите $\MSE$ для этой оценки. Является ли оценка $\mletheta$ состоятельной?
1. Найдите с помощью метода моментов оценку $\theta_{M}$ для $\theta$. Найдите предельное распределение для случайной величины
	$$
	\sqrt{n}(\theta_{M} - \theta).
	$$

---

1.

$\displaystyle{
\begin{align*}
\theta_{MLE} =
& \argmax_\theta \sum_i \ln p(X_i|\theta) =\\=
& \argmax_\theta \sum_i \ln \frac{1}{\theta}\mathbb{1}[X_i \in [\theta, 2\theta]] =\\=
& \arg\max_\theta \left( -n\ln \theta + \sum_i \ln \mathbb{1}[X_i \in [\theta, 2\theta]]\right) =\\=
& \arg\max_\theta \left( -n\ln \theta - \infty\sum_i \mathbb{1}[X_i \not\in [\theta, 2\theta]]\right)
\end{align*}
}$

То есть максимум правдоподобия достигается в минимальном $\theta$, при условии, что $0 < \theta \leq \boldX_{(1)}$ и $\boldX_{(n)} \leq 2\theta$, получаем $\theta_{MLE} = \cfrac{\boldX_{(n)}}{2}$

$\displaystyle{
\begin{align*}
\text{MSE}(\theta_{MLE}) = 
\Exp((\theta_{MLE} - \theta)^2) =
\frac{1}{4}\Exp(\boldX_{(n)}^2) -\Exp(\boldX_{(n)})\theta +  \theta^2
\end{align*}
}$

$\displaystyle{
F(x)_{\boldX_{(n)}} =
P(\boldX_{(n)} < x) =
\prod_i P(\boldX_i < x) =
\prod_i \frac{x - \theta}{\theta} =
\left(\frac{x - \theta}{\theta}\right)^n
}$

$\displaystyle{
\begin{align*}
\Exp(\boldX_{(n)}) =
& \int_\theta^{2\theta} x dF(x)_{\boldX_{(n)}} =\\=
& \int_\theta^{2\theta} x\frac{n(x - \theta)^{n-1}}{\theta^n}dx =\\=
& \int_0^{\theta} (y + \theta)\frac{ny^{n-1}}{\theta^n}dy =\\=
& \frac{n}{\theta^n}\int_0^{\theta}\left( y^n + \theta y^{n-1}\right)dy =\\=
& \frac{n}{\theta^n}\left( \frac{y^{n+1}}{n+1} + \theta \frac{y^n}{n}\right)\big|_0^\theta =\\=
& \frac{n}{\theta^n}\left( \frac{\theta^{n+1}}{n+1} + \frac{\theta^{n+1}}{n}\right) =\\=
& \theta\frac{2n+1}{n+1}
\end{align*}
}$

$\displaystyle{
\begin{align*}
\Exp(\boldX_{(n)}^2) =
& \int_\theta^{2\theta} x^2 dF(x)_{\boldX_{(n)}} =\\=
& \int_0^{\theta} (y + \theta)^2\frac{ny^{n-1}}{\theta^n}dy =\\=
& \frac{n}{\theta^n}\int_0^{\theta}\left( y^{n+1} + 2\theta y^{n} + \theta^2y^{n-1}\right)dy =\\=
& \frac{n}{\theta^n}\left( \frac{y^{n+2}}{n+2} + 2\theta \frac{y^{n+1}}{n+1} + 
\theta^2\frac{y^{n}}{n}\right)\big|_0^\theta =\\=
& \frac{n}{\theta^n}\left( \frac{\theta^{n+2}}{n+2} + 2\frac{\theta^{n+2}}{n+1} + 
\frac{\theta^{n+2}}{n}\right) =\\=
& \theta^2\left( \frac{n}{n+2} + 2\frac{n}{n+1} + 1\right)
\end{align*}
}$

$\displaystyle{
\text{MSE}(\theta_{MLE}) =
\frac{1}{4} \theta^2\left( \frac{n}{n+2} + 2\frac{n}{n+1} + 1\right) - \theta^2\frac{2n+1}{n+1} + \theta^2 =
\frac{\theta^2}{2(n+1)(n+2)} \xrightarrow[n \to \infty]{} 0,
}$

Значит, оценка состоятельна.

2.

$\displaystyle{
\Exp[X] = \frac{3}{2}\theta \\
\mean{\boldX} = \frac{3}{2}\theta_{M}\\
\theta_{M} = \frac{2}{3}\mean{\boldX}
}$

Заметим, что дисперсия $X$ равна $\sigma^2 = \cfrac{(2\theta - \theta)^2}{12} = \cfrac{\theta^2}{12}$, значит согласно центральной предельной теореме,

$\displaystyle{
\frac{\sqrt n }{\sigma} (\mean\boldX - \Exp[X]) \xrightarrow[]{\mathcal{D}} \Normal(0,1) \\
\frac{\sqrt n }{\sigma} \left(\frac{3}{2}\theta_M - \frac{3}{2}\theta\right) \xrightarrow[]{\mathcal{D}} \Normal(0,1) \\
\sqrt n \left(\theta_M - \theta\right) \xrightarrow[]{\mathcal{D}} \Normal\left(0,\frac{4\sigma^2}{9}\right) =
\Normal\left(0,\frac{\theta^2}{27}\right)\\
}$

Где последний переход совершен согласно дельта-методу.

### Задача 5
Пусть $\boldX = \{X_1,\ldots,X_n\} $ –– н.о.р. случайные величины, имеющие мультиномиальное распределение с параметром $\boldp = (p_1, p_2, p_3)^T$, т.е. $X_i\in\{1,2,3\}$ и

	$$
	\psi = \log p_1.
	$$

---

1.

$\displaystyle{
L(\boldp) = \prod_i \prod_j p_j ^{\mathbb{1}[X_i == j]} \\
l(\boldp) = \sum_i \sum_j \mathbb{1}[X_i == j] \ln p_j  \\

Для нахождения максимума лог-правдоподобия с ограничением $\sum_j p_j = 1$, применим метод множителей Лагранжа:

$\displaystyle{
\hat \boldp = 
\arg\max \left(l(\boldp) + \lambda(\sum_j p_j - 1)\right) =
\arg\max \left(\sum_i \sum_j \mathbb{1}[X_i == j] \ln p_j + \lambda(\sum_j p_j - 1)\right)
}$

$\displaystyle{
\frac{d}{d p_j} = \sum_i \mathbb{1}[X_i == j] \frac{1}{p_j} - \lambda\\
\hat p_j = \frac{\sum_i \mathbb{1}[X_i == j]}{\lambda} \\
\sum_j \hat p_j = \frac{\sum_j\sum_i \mathbb{1}[X_i == j]}{\lambda} = \frac{n}{\lambda} = 1 \\
\hat p_j = \frac{\sum_i \mathbb{1}[X_i == j]}{n}
}$

$\displaystyle{
\begin{align*}
[{\mathcal {I}}(\boldp )]_{i,j}=
& -\Exp \left[\left.{\frac {\partial ^{2}}{\partial p_{i}\,\partial p_{j}}}\log f(\boldX;\boldp )\right|\boldp \right] =\\=
& -\Exp \left[\left.{\frac {\partial ^{2}}{\partial p_{i}\,\partial p_{j}}}\sum_k \sum_l \mathbb{1}[X_k == l] \ln p_l\right|\boldp \right] =\\=
& -\Exp \left[\left.{\frac {\partial}{\partial p_{i}}}\sum_k \mathbb{1}[X_k == j] \frac{1}{p_j}\right|\boldp \right] =\\=
& \Exp \left[\left.\mathbb{1}[i == j]\sum_k \mathbb{1}[X_k == j] \frac{1}{p_j^2}\right|\boldp \right] =\\=
& \frac{1}{p_j^2}\mathbb{1}[i == j] \sum_k \Exp \left[\left. \mathbb{1}[X_k == j] \right|\boldp \right] =\\=
& \frac{1}{p_j^2}\mathbb{1}[i == j] \sum_k p_j =\\=
& \frac{n}{p_j}\mathbb{1}[i == j]
\end{align*}
}$

Для оценки максимального правдоподобия в условиях регулярности выполняется следующее:

$\displaystyle{
\sqrt n (\hat\boldp - \boldp) \xrightarrow[]{\mathcal{D}} \Normal(0, {\mathcal {I_0}}(\boldp )^{-1}) = 
\Normal(0, n{\mathcal {I}}(\boldp )^{-1}) \\

Заметим, что истинную матрицу Фишера мы не знаем, но в условиях регулярности и в пределе данное выражение будет верным, если заменить ${\mathcal {I}}(\boldp )$ на ${\mathcal {I}}(\hat\boldp )$. Далее, поскольку матрица ковариаций ${\mathcal {I}}(\hat\boldp)$ диагональна, то

$\displaystyle{
