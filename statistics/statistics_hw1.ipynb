## Task 1 (2)<sup>[](#_toc)</sup>

Пусть $\boldX = \{X_1,\ldots, X_n\} \sim \Uniform(0,\theta)$, $\esttheta = 2\mean{\boldX}$. Найдите значения значения $\bias$, $\se$ и $\MSE$ этой оценки. Является ли оценка несмещенной? Состоятельной?

---
1.
$
\text{bias} = 
\Exp(\hattheta) - \theta =
\Exp(2\mean\boldX) - \theta =
\frac{2}{n}\sum_i(\Exp X_i) - \theta =
\frac{2}{n}\frac{n\theta}{2} - \theta = 0
$

2.
$\text{se} = 
\sqrt{\Var(\hattheta)} =
\sqrt{\Exp(\hattheta^2) -\Exp(\hattheta)^2} =
\sqrt{\Exp((2\mean\boldX)^2) - \theta^2} =
\sqrt{\frac{4}{n^2}\Exp(\sum_i X_i^2 + \sum_{i\neq j} X_iX_j) - \theta^2} =$

$
= \sqrt{\frac{4}{n^2}n\Exp(X_1^2) + \frac{4}{n^2}n(n-1)\Exp(X_1)^2 - \theta^2} =
\sqrt{\frac{4}{n}\Exp(X_1^2) + \frac{(n-1)\theta^2}{n} - \theta^2}\\
$

Учитывая, что 
$\Exp(X_1^2) = 
\int_0^1 x^2 dF(x) = 
\int_0^{\theta} x^2 \cfrac{1}{\theta} dx =
\cfrac{x^3}{3\theta} \Big|_0^{\theta} =
\cfrac{\theta^2}{3}
$

Получаем:

$
se = \sqrt{\frac{4}{n}\frac{\theta^2}{3} + \frac{(n-1)\theta^2}{n} - \theta^2} = \frac{\theta}{\sqrt{3n}}
$

3.
$\text{MSE} = 
\Exp(\hattheta - \theta)^2 =
\Exp(\hattheta^2) -2\theta\Exp(\hattheta) +\theta^2 =
\Exp(\hattheta^2) -\Exp(\hattheta) ^2 + (\Exp(\hattheta) - \theta)^2=
\text{var} + \text{bias}^2 = 
\text{se}^2 + \text{bias}^2 = 
\frac{\theta^2}{3n}
$

Поскольку bias равен нулю, то оценка $\hattheta$ параметра $\theta$ является несмещенной, а так как, кроме того, ее дисперсия стремится к 0 с ростом числа наблюдений (числа независимых одинаково распределенных случайных величин $X_i$), то она также является состоятельной.

## Task 2 (3)

Пусть $\boldX = \{X_1, \dots, X_n\} \sim \Poisson(\ln \lambda)$, $\estlambda = e^{\mean{\boldX}}$. Найдите $\bias$, $\se$ и $\MSE$ этой оценки. Является ли оценка несмещенной? Состоятельной?

---

1.
$\text{bias} = 
\Exp(\hat\lambda) - \lambda =
\Exp(e^{\mean\boldX}) - \lambda =
\Exp(e^{\sum_i X_i / n}) - \lambda =
\prod_i\Exp(e^{X_i / n}) - \lambda =
(\int e^{x / n} dF(x))^n - \lambda =
(\sum_{x=0}^\infty e^{x / n} \frac{(\ln \lambda)^x}{\lambda x!} )^n - \lambda =
$

$=
(\sum_{x=0}^\infty  \frac{(e^{1 / n} \ln \lambda)^x}{\lambda x!} )^n - \lambda =
(\frac{\exp{(e^{1 / n} \ln \lambda)}}{\lambda})^n - \lambda =
(\lambda ^{e^{1 / n} - 1})^n - \lambda =
\lambda ^{n(e^{1 / n} - 1)} - \lambda
$

<br>

$\lim_{n\rightarrow \infty} \text{bias} = 
\lim_{n\rightarrow \infty} \lambda ^{n(e^{1 / n} - 1)} - \lambda =
\lim_{n\rightarrow \infty} \lambda ^{n(1 + 1 / n + o(1/n) - 1)} - \lambda =
\lim_{n\rightarrow \infty} \lambda ^{1 + o(1)} - \lambda = 0
$

<br>

2.
$\text{se} = 
\sqrt{\Var(\hat\lambda)} =
\sqrt{\Exp(\hat\lambda^2) -\Exp(\hat\lambda)^2} =
\sqrt{\Exp(e^{2 \sum_i X_i / n}) -\Exp(\hat\lambda)^2} =
\sqrt{\lambda ^{n(e^{2 / n} - 1)} -\lambda ^{2n(e^{1 / n} - 1)}}
$

<br>


$\lim_{n\rightarrow \infty} \text{se} = 
\lim_{n\rightarrow \infty} \sqrt{\lambda ^{n(e^{2 / n} - 1)} -\lambda ^{2n(e^{1 / n} - 1)}} =
\lim_{n\rightarrow \infty} \sqrt{\lambda ^{n(2 / n + 2/n^2 +  o(1/n^2))} -\lambda ^{2n(1/n + 1/2n^2+ o(1/n^2))}} =
\lim_{n\rightarrow \infty} \sqrt{\lambda ^{2 + 2/n +  o(1/n))} -\lambda ^{2 + 1/n+ o(1/n)}} =\\=
\lim_{n\rightarrow \infty} \lambda \sqrt{ \lambda^{2/n +  o(1/n)} -\lambda ^{ 1/n+ o(1/n)}} =
\lim_{n\rightarrow \infty} \lambda \sqrt{ \frac{2\ln\lambda}{n} +  o(\frac{1}{n}) - \frac{\ln\lambda}{n}+ o(\frac{1}{n})} =
\lim_{n\rightarrow \infty} \lambda \sqrt{\frac{\ln\lambda}{n}+ o(\frac{1}{n})} = 0
$

<br>

3.
$\text{MSE} = 
\text{var} + \text{bias}^2 = 
\text{se}^2 + \text{bias}^2 = 
\lambda ^{n(e^{2 / n} - 1)} -\lambda ^{2n(e^{1 / n} - 1)} + (\lambda ^{n(e^{1 / n} - 1)} - \lambda)^2 =
\lambda ^{n(e^{2 / n} - 1)}- 2\lambda ^{n(e^{1 / n} - 1) + 1} + \lambda^2
$

Эта оценка является смещенной, но поскольку она асимптотически несмещенная и ее дисперсия стремится к нулю, она также является состоятельной.

## Task 5 (2)<sup>[toc](#_toc)</sup>

Скачайте данные о качестве красных вин. Постройте график для $\hatF(x;\boldX)$ для уровня кислотности (pH). Для каждой точки $x$ постройте:
* 95\%-ый доверительный интервал на основе неравенства Дворецкого-Кифера-Вольфовица.
* Асимптотический нормальный 95\%-ый доверительный интервал для значения $F(x)$.

По значениям уровня кислотности $\boldX$ подсчитайте оценку $T(\boldX)$ для функционала $T(F) = F(3.5) - F(3.4)$ и найдите оцените аналитически стандартное отклонение $\estSe$ оценки $T(\boldX)$. Постройте асимптотический нормальный 95\%-ый доверительный интервал для $T(F)$.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

from scipy import stats
```

```python
# Предполагается, что датасет лежит по соседству с ноутбуком
phs = pd.read_csv(os.path.join(os.curdir, "winequality-red.csv"), sep=";")
X = phs["pH"].values
LEFT = 3.4
RIGHT = 3.5
print("Number of samples: {}".format(X.shape))
print("Min value: {}".format(X.min()))
print("Max value: {}".format(X.max()))
```

### Solution<sup>[toc](#_toc)</sup>

```python
class SampleStats:
    def __init__(self, samples):
        self.sorted_samples = np.sort(samples)
        self.n = len(samples)
        self.linspace = np.linspace(
            self.sorted_samples[0], self.sorted_samples[-1], self.n
        )

    def empirical_distribution_function(self, x):
        return np.searchsorted(self.sorted_samples, x) / self.n

    edf = empirical_distribution_function

    def print_edf(self):
        plt.plot(
            self.linspace,
            [self.empirical_distribution_function(x) for x in self.linspace],
            label="edf",
        )
        plt.legend()

    def get_dvoretzky_kiefer_wolfowitz_confidence_band_width(
        self, confidence_level=0.95
    ):
        alpha = 1 - confidence_level
        return (np.log(2 / alpha) / (2 * self.n)) ** 0.5

    def get_asymptoticaly_normal_confidence_band_width(self, confidence_level=0.95):
        alpha = 1 - confidence_level
        return stats.norm().ppf(1 - alpha / 2) * stats.sem(self.sorted_samples)

    def print_confidence_interval(self, confidence_band_width, name):
        lower, upper = [
            [
                min(
                    1,
                    max(
                        0,
                        self.empirical_distribution_function(x)
                        + i * confidence_band_width,
                    ),
                )
                for x in self.linspace
            ]
            for i in [-1, 1]
        ]
        plt_kwargs = dict(alpha=0.95, ls="dotted")
        (ax,) = plt.plot(self.linspace, lower, label=name, **plt_kwargs)
        plt.plot(self.linspace, upper, c=ax.get_color(), **plt_kwargs)
        plt.legend()
```

```python
ph_stats = SampleStats(phs["pH"])

ph_stats.print_edf()
ph_stats.print_confidence_interval(
    ph_stats.get_dvoretzky_kiefer_wolfowitz_confidence_band_width(),
    "Dvoretzky Kiefer Wolfowitz",
)
ph_stats.print_confidence_interval(
    ph_stats.get_asymptoticaly_normal_confidence_band_width(), "Asymptoticaly normal"
)
```

```python
def get_asymptoticaly_normal_confidence_interval(
    mean, standard_error_of_the_mean, confidence_level=0.95
):
    alpha = 1 - confidence_level
    return [
        mean + i * stats.norm().ppf(1 - alpha / 2) * standard_error_of_the_mean
        for i in [-1, 1]
    ]
```

```python
estimate = ph_stats.empirical_distribution_function(
    RIGHT
) - ph_stats.empirical_distribution_function(LEFT)
standard_error = (estimate * (1 - estimate) / ph_stats.n) ** 0.5

interval = get_asymptoticaly_normal_confidence_interval(estimate, standard_error)
```

```python
plt.title("95% Asymptoticaly normal confidence interval for functional")

x = 0
plt.plot([x, x], interval, "o-")
plt.plot(x, estimate, "o")
```

## Task 7 (2)<sup>[toc](#_toc)</sup>

Провести моделирование, чтобы сравнить различные типы доверительных интервалов, построенных с помощью бутстрепа. Пусть $n = 50$, $T(F) = {\int\lp x-\mu\rp^3dF(x)} / \sigma^{3}$ &mdash; коэффициент асимметрии, где $F$ &mdash; распределение $\chi_k^2$ с тремя степенями свободы ($k = 3$). Постройте 95\% доверительные интервалы для $T(F)$ по выборке $\boldX = \{X_1,\ldots,X_n\}$, используя три подхода на основе бутстрепа.

### Solution<sup>[toc](#_toc)</sup>

```python
n_samples = 50
chi2_degrees_of_freedom = 3

chi2 = stats.chi2(df=chi2_degrees_of_freedom)
chi2_sample = chi2.rvs(n_samples)

plt.hist(chi2_sample, bins=30, label="sample distribution")
plt.legend();
```

```python
class AssymetryCoefficientFunctional:
    def __init__(self, mean, std):
        self.mean = mean
        self.std = std

    def __call__(self, x):
        return (x - self.mean) ** 3 / self.std**3


true_functional_value = chi2.expect(
    AssymetryCoefficientFunctional(mean=chi2.mean(), std=chi2.std())
)


def assymetry_coefficient_estimate(sample):
    functional = AssymetryCoefficientFunctional(mean=sample.mean(), std=sample.std())
    return functional(sample).mean()


estimate = assymetry_coefficient_estimate(chi2_sample)


def bootstrap(sample):
    return np.random.choice(sample, len(sample))


def generate_bootstrap_samples(sample, n_bootstrap_samples):
    return np.array([bootstrap(sample) for i in range(n_bootstrap_samples)])


def calculate_bootstrap_estimates(bootstrap_samples, sample_estimate_callable):
    return np.array(
        [
            sample_estimate_callable(bootstrap_sample)
            for bootstrap_sample in bootstrap_samples
        ]
    )


bootstrap_estimates = calculate_bootstrap_estimates(
    bootstrap_samples=generate_bootstrap_samples(
        sample=chi2_sample, n_bootstrap_samples=1000
    ),
    sample_estimate_callable=assymetry_coefficient_estimate,
)


def get_central_confidence_interval(bootstrap_estimates, confidence_level=0.95):
    alpha = 1 - confidence_level
    estimate = bootstrap_estimates.mean()
    return [
        2 * estimate - np.quantile(bootstrap_estimates, 1 - alpha / 2),
        2 * estimate - np.quantile(bootstrap_estimates, alpha / 2),
    ]


def get_percentile_confidence_interval(bootstrap_estimates, confidence_level=0.95):
    alpha = 1 - confidence_level
    return [
        np.quantile(bootstrap_estimates, alpha / 2),
        np.quantile(bootstrap_estimates, 1 - alpha / 2),
    ]
```

```python
def plot_three_different_confidence_intervals(
    true_functional_value,
    estimate,
    bootstrap_estimates,
    confidence_level=0.95,
    interval_y_level=0.04,
):
    asymptoticaly_normal_confidence_interval = (
        get_asymptoticaly_normal_confidence_interval(
            mean=estimate,
            standard_error_of_the_mean=bootstrap_estimates.std(),
            confidence_level=confidence_level,
        )
    )

    central_confidence_interval = get_central_confidence_interval(
        bootstrap_estimates, confidence_level
    )

    percentile_confidence_interval = get_percentile_confidence_interval(
        bootstrap_estimates, confidence_level
    )

    plt.title("Different methods 95% confidence intervals")
    plt.hist(
        bootstrap_estimates,
        bins=30,
        density=True,
        label="bootstrap functional estimates",
        histtype="step",
    )
    plt.axvline(true_functional_value, ls="--", c="grey", label="true functional value")
    plt.axvline(estimate, c="black", label="estimate functional value")

    interval_y = np.repeat(interval_y_level, 2)
    plt.plot(
        asymptoticaly_normal_confidence_interval,
        interval_y,
        lw=7,
        label="asymptoticaly normal",
    )
    plt.plot(central_confidence_interval, 2 * interval_y, lw=7, label="central")
    plt.plot(percentile_confidence_interval, 3 * interval_y, lw=7, label="percentile")

    plt.legend();
```

```python
plot_three_different_confidence_intervals(
    true_functional_value, estimate, bootstrap_estimates
)
```

## Task 9 (2)<sup>[toc](#_toc)</sup>

Пусть $\boldX = \{X_1,\ldots, X_n\} \sim \Exponential(\lambda)$, $\theta = e^{\frac{1}{\lambda}}$ и $\esttheta = e^{\mean{\boldX}}$. Сгенерируйте выборку $\boldX$ из $n = 1000$ наблюдений для $\lambda = 0.25$. Нарисуйте гистограмму значений $\{\esttheta^*_i\}_{i=1}^B$ бутстрепных оценок. Эта гистограмма является оценкой распределения $p_{\esttheta}(x)$. Сравните ее с настоящим распределением $p_{\esttheta}(x)$, вычисленным аналитически в предыдущем пункте. Используя бутстреп, подсчитайте величину $\se$ и постройте тремя способами $95\%$-ый доверительный интервал для $\theta$. 

### Solution<sup>[toc](#_toc)</sup>

```python
exp_lambda = 0.25
sample_size = 1000
true_theta = np.e ** (1 / exp_lambda)

exp = stats.expon(scale=1 / exp_lambda)
exp_sample = exp.rvs(sample_size)
```

$p_{\esttheta}(x) = 
\frac{(\lambda n)^{n}}{(n-1)!} \frac{(\ln x)^{n-1}}{x^{\lambda n + 1}} =
\lambda n\frac{(\lambda n)^{n-1}}{(n-1)!} \frac{1}{x^{\lambda + 1}} (\frac{\ln x}{x^{\lambda }})^{n-1} :=
c \frac{1}{x^{\lambda + 1}} (\frac{\ln x}{x^{\lambda }})^{n-1}
$

```python
class AnalyticalEstimatePdf:
    def __init__(self, exp_lambda, n_samples):
        self.exp_lambda = exp_lambda
        self.n = n_samples

        self.constant = self.exp_lambda * self.n
        for i in range(1, self.n):
            self.constant *= self.exp_lambda * self.n / i

    def __call__(self, x):
        x[x < 1] = 1
        return (
            self.constant
            / x ** (self.exp_lambda + 1)
            * (np.log(x) / x**self.exp_lambda) ** (self.n - 1)
        )


pdf = AnalyticalEstimatePdf(exp_lambda=exp_lambda, n_samples=sample_size)
```

```python
def theta_estimate(sample):
    return np.e ** (sample.mean())


exp_bootstrap_estimates = calculate_bootstrap_estimates(
    bootstrap_samples=generate_bootstrap_samples(
        sample=exp_sample, n_bootstrap_samples=1000
    ),
    sample_estimate_callable=theta_estimate,
)
```

```python
plot_range = [30, 80]

linspace = np.linspace(*plot_range, num=1000)
plt.hist(
    exp_bootstrap_estimates,
    bins=30,
    density=True,
    histtype="step",
    label="estimate pdf",
)
plt.plot(linspace, pdf(linspace), label="analytical pdf")
plt.legend();
```
