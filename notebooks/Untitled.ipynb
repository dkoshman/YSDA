```python
class Solution:
    def findWords(self, board, words):
        self.words = words
        self.board = board
        self.found_words = set()
        self.m = len(board)
        self.n = len(board[0])
        for row in range(self.m):
            for col in range(self.n):
                self.find(row, col, set(), self.words)
        return list(self.found_words)
        
    def find(self, row, col, path, candidates):
        if not candidates:
            return
        
        size = len(path)
        letter = self.board[row][col]
        new_candidates = []
        print(row, col)
        for candidate in candidates:
            if candidate not in self.found_words:
                if candidate[size] == letter:
                    if len(candidate) == size + 1:
                        self.found_words.add(candidate)
                    else:
                        new_candidates.append(candidate)
        
        for r, c in [(row - 1, col), (row, col - 1), (row, col + 1), (row + 1, col)]:
            if 0 <= r < self.m and 0 <= c < self.n and (r, c) not in path:
                self.find(r, c, path | {(row, col)}, new_candidates)
        
```
