```python
#first class
def gen_data(seed,n = 100):
    np.random.seed(seed)
    data = Namespace()
    data.seed = seed
    data.x = np.random.normal(size=(n, 2))
    data.y = np.zeros(n)

    #add second class
    data.x = np.r_[data.x, np.random.normal(size=(n, 2), loc=2)]
    data.y = np.r_[data.y, np.ones(n)]
    
    return data


data_42 = gen_data(seed=42, n = 100)

plt.figure(figsize=(10.0, 6.0))
plt.scatter(data_42.x[:, 0], data_42.x[:, 1], c=data_42.y, s=100, cmap='autumn')
plt.axline((0,2), (2,0))
```

```python
def get_grid(data):
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01),
                    np.arange(y_min, y_max, 0.01))



def plot_model(X, y, clf, proba=False):
    clf.fit(X, y)
    xx, yy = get_grid(X)
    if proba:
        predicted = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)
    else:
        predicted = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.pcolormesh(xx, yy, predicted, cmap='autumn', shading='auto',alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=150, cmap='autumn', alpha=0.7,edgecolors="black")
    plt.ylim([yy.min(),yy.max()])
    plt.xlim([xx.min(),xx.max()])
    return clf
```

```python
def get_common_grid(data):
    x_min = 0
    x_max = 0
    y_min = 0
    y_max = 0
    for the_data in data:
        data_x_min, data_x_max = the_data.x[:, 0].min() - 1, the_data.x[:, 0].max() + 1
        data_y_min, data_y_max = the_data.x[:, 1].min() - 1, the_data.x[:, 1].max() + 1
        x_min = min(x_min, data_x_min)
        x_max = max(x_max, data_x_max)

        y_min = min(y_min, data_y_min)
        y_max = max(y_max, data_y_max)
        
    return np.meshgrid(np.arange(x_min, x_max, 0.01),
                    np.arange(y_min, y_max, 0.01))

def plot_multi_models(data, clf, proba=False, show_std=False):
    xx, yy = get_common_grid(data)

    predicted_stack = []
    
    for the_data in data:
        clf.fit(the_data.x, the_data.y)
        if proba:
            predicted = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)
        else:
            predicted = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
            
        predicted_stack.append(predicted)
        
    predicted_stack = np.stack(predicted_stack,axis=0)
    if show_std:
        mean_predicted_stack = predicted_stack.std(axis=0)
    else:
        mean_predicted_stack = predicted_stack.mean(axis=0)
        
    plt.pcolormesh(xx, yy, mean_predicted_stack, cmap='autumn', shading='auto',alpha=0.5)
    #plt.scatter(X[:, 0], X[:, 1], c=y, s=150, cmap='autumn', alpha=0.7,edgecolors="black")
    plt.ylim([yy.min(),yy.max()])
    plt.xlim([xx.min(),xx.max()])
```

```python
plt.figure(figsize=(10.0, 6.0))
plot_multi_models([data_42, data_43], DecisionTreeClassifier(min_samples_leaf=5), proba=True)
plt.axline((0,2), (2,0))

plt.figure(figsize=(10.0, 6.0))
plot_multi_models([data_42, data_43], DecisionTreeClassifier(min_samples_leaf=5), proba=True, show_std=True)
plt.axline((0,2), (2,0))
```

А формально как это понять?

Будучи обученной на этом множестве класификатор в виде решающего дерева это функция от сл. величины, т.е. тоже сл. величина.

Теперь у нас есть еще усреднение от деревьев, по бутстрапированным выборка построенным.
(кстати, давайте поёмём, что это значит)

это значит, что выбросы могут не попасть при обучении. они будут влиять только на 63% в ансабле.
т.е. если рядом нет точек, которые "локально доопишут пространство", если "выброс" не попал в выборку, то искажение будет тише в общем ансамбле.

(но, верно и обратное, когда-то у выброса будет вес 1, или больше)

## Сходимость моделей

```python
from sklearn.metrics import mean_squared_error as mse

def get_mse_list(X, y, clf, print_train=True, skip=0):
    predictions = list()
    mses = list()
    for i, tree in enumerate(clf.estimators_):
        if i < skip:
            continue
        predictions.append(tree.predict(X))
        rf_prediction = np.mean(predictions, axis=0)
        mses.append(mse(y, rf_prediction))
    if print_train:
        print('Last score', mses[-1])
    return mses
```

```python
def plot_mse_graph(clf, X_train, X_test, y_train, y_test, prefix, print_train=True, skip=0):
    clf.fit(X_train, y_train)
    train_mses = get_mse_list(X_train, y_train, clf, False, skip=skip)
    test_mses = get_mse_list(X_test, y_test, clf, skip=skip)
    if print_train:
        plt.plot(train_mses, label=prefix + ' train set error')
    plt.plot(test_mses, label=prefix + ' test set error')
    plt.xlabel('n_estimators')
    plt.ylabel('MSE')
    plt.legend()
```

```python
%%time
n_estimators = 50
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor

plt.figure(figsize=(10.0, 6.0))
plot_mse_graph(RandomForestRegressor(n_estimators=n_estimators, n_jobs=5, random_state=0),
               X_train, X_test, y_train, y_test,
               prefix='RF')
plot_mse_graph(ExtraTreesRegressor(n_estimators=n_estimators, n_jobs=5, bootstrap=True, random_state=0),
               X_train, X_test, y_train, y_test,
               prefix='ET')

plot_mse_graph(BaggingRegressor(
    base_estimator=DecisionTreeRegressor(),
    n_estimators=n_estimators, random_state=0, n_jobs=3),
               X_train, X_test, y_train, y_test,
               prefix='BA_DTR')

plt.grid()
plt.show()
```

### Что такое extremely randomized trees:

Идея: еще более нескорелированные члены в ансамбле.
  * Экономим на ресамблинге данных.
  * Тоже берём подмножество признаков при построении сплита, но берем не лучший, а несколько случайных, и утверждаем тот, что больше других уменьшает суммарную нечистоту после разбиения.

## Какие гиперпараметры настраивать для RandomForest

Дьяконов в статье https://dyakonov.org/2016/11/14/случайный-лес-random-forest/ выделяет
  * Параметры самого RF
    * n_estimators  -- в какой-то момент качество на отложенной выборке/оценке по CV стабилизируется. Больше будет неэффективно с т.з. cpu
    * max_features -- можно пытаться понизить скорелированность разных деревьев
  * парамтеры дерева решений
    * max_depth  -- на максимум, если позволяет память
